```python
df = pd.read_csv('./AllFlightsGrouped.csv', header = 0)
df.head()
```

```python
data = pd.DataFrame(columns=['ORIGIN', 'DEST', 'YEAR', 'ON_TIME', 'DELAYED15MIN', 'DIVERTED', 'CANCELLED', 'TOTAL'])
```

```python
LABELS = ['ON_TIME', 'DELAYED15MIN', 'DIVERTED', 'CANCELLED', 'TOTAL']
for label in LABELS:
    df[label] = 0
```

```python
def count_types(x):
    count = x['MONTH']
    label_index = x['LABEL']
    x[LABELS[label_index]] = count
    x['TOTAL'] = count
    return x
    
df = df.apply(count_types, axis = 1)
```

```python
df.head()
```

```python
data = df.groupby(by = ['ORIGIN', 'DEST']).sum()
```

```python
df2 = data.loc[:, data.columns.isin(LABELS + ['ORIGIN', 'DEST'])]
```

```python
df2.head()
```

```python
df2.to_csv('./summary_all_flights.csv', header = True, index = True)
```
